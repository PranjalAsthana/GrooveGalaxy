#Vectorization

```python
import nltk
nltk.download('punkt')
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
def stemfunc(txt):
  token = nltk.word_tokenize(txt)
  a = [stemmer.stem(w) for w in token]
  return " ".join(a)
```

```python
stemfunc("this is so beautiful it's amazing yet I just cant believe what I am seeing")
```

```python
#songs['text'].apply(lambda x: stemfunc(x))
```

```python
songs['text'] = songs['text'].apply(stemfunc)
```

#TF-IDF Vectorisation

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(analyzer='word', stop_words='english')
```

```python
matrix = tfidf.fit_transform(songs['text'])
```

#Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(matrix)
```

```python
similarity[0]
```

#Recommender Function

```python
def recommender(song_name):
  idx = songs[songs['song']==song_name].index[0]
  distance = sorted(list(enumerate(similarity[idx])), reverse=True, key = lambda x:x[1])
  song = []
  for s in distance[1:6]:
    song.append(songs.iloc[s[0]].song)
  return song
```
